# Excersise on various Visualization Techniques

The iris dataset contains measurements for 150 iris flowers from three different species.


The three classes in the Iris dataset:
Iris-setosa (n=50)
Iris-versicolor (n=50)
Iris-virginica (n=50)
The four features of the Iris dataset:
sepal length in cm
sepal width in cm
petal length in cm
petal width in cm


Get the data
*Use seaborn to get the iris data by using: iris = sns.load_dataset('iris') *



```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(data=iris,hue="species")
```

Create a kde plot of sepal_length versus sepal width for setosa species of flower.

```python
setosa = iris[iris['species']=='setosa']
sns.kdeplot( setosa['sepal_width'], setosa['sepal_length'],cmap="plasma", shade=True, shade_lowest=False)
```

```python
tips=sns.load_dataset('tips')
tips['tip_pct']=tips['tip']*100/tips['total_bill']
graph=sns.FacetGrid(data=tips,row='sex',col='time')
graph.map(sns.distplot,'tip_pct',color='blue',bins=10,kde=False)
```

```python
tips=sns.load_dataset('tips')
sns.catplot(x="day", y="total_bill", hue="sex", kind="box", data=tips);
```

```python
tips=sns.load_dataset('tips')
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind ="hex",color="lightblue")
```

```python
a=np.arange(11)
plt.plot(a,a,label="linear")
plt.legend()
```

```python
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.plot(x,z,linestyle="dashed")
```

```python
x=np.arange(11)

fig, axes = plt.subplots(nrows=1,ncols= 2)
      
axes[0].plot(x, x**2,x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Log scale");
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
a=np.linspace(-0.75,1,100)
plt.plot(a,a**3)
```
